```python
def multi_predict(x_values :pd.DataFrame, models) -> pd.DataFrame:
    """
    Function for predicting on multiple models and averaging the results
    """
    results = models[0].predict(x_values)
    for model in models[1:]:
        prediction = model.predict(x_values)
        results += prediction
    
    results = results / len(models)

    return results
```

```python
def average_predictions(x_values: pd.DataFrame, models_2d_list) -> pd.DataFrame:
    """
    Function for predicting on multiple sets of LightGBM models,
    averaging the predictions within each set, and then averaging
    the results across all sets.
    """
    all_results = []

    # Iterate over the 2D list and call multi_predict() for each inner list
    for models in models_2d_list:
        result = multi_predict(x_values, models)
        all_results.append(result)

    # Average the results from each call to multi_predict()
    avg_result = np.mean(all_results, axis=0)

    return pd.DataFrame(avg_result)
```
